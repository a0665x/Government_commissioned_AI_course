```python
# my_data_list = [1,2,3,4,5,6,7]
def cal_mean_std(my_data_list,output_type ):  #我這個定義吃的型態是list  output_type='mean' /'std'
    sum_t = 0
    for i in my_data_list:
        sum_t = sum_t+i
    mean = sum_t/len(my_data_list)
    #開始算標準差
    sd = 0
    for j in my_data_list:
        sd = sd + (j-mean)**2
    std = (sd/(len(my_data_list)-1))**0.5
    if output_type=='mean':
        return mean
    elif output_type=='std':
        return std
```

```python
data = [10,2,3,57,45,2]
mean = cal_mean_std(data,'mean')
std = cal_mean_std(data,'std')
print(f'數據的平均值:{mean}')
print(f'數據的標準差:{std}')
```

```python
# def 輸入一個資料 以及一個閥值 去判斷資料是否超出閥值以及超出閥值的平均  
#範例: input  =[10,2,3,57,45,2] TH=50 output=[57,45] 以及平均
def data_filter(data_list, TH):
    filter_list = []
    for i in data_list:
        if i >TH:
            filter_list.append(i) 
    #東西已經經過過濾完了
    mean = cal_mean_std(filter_list , 'mean')
    return filter_list , mean
```

```python
data = [1,2,7,8,4,3,7,9,5,8,9,4,3,2,6]
filter_list , mean = data_filter(data,5)
print('數據的閥值是5過濾後:',filter_list)
print('數據的閥值是5過濾後的平均:',mean)
```
