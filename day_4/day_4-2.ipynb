```python
# class 用途: 當今天我們設計的def 非常多的時候, 需要一個更高級別的封包, class 即可方便的管理所寫的def
# ex:數值計算
class my_cla_funs:
    def __init__(self,value1,value2): #做額外變數設定
        self.p11 = value1
        self.p22 = value2
        
    def plus(self):
        _sum_ = self.p1+self.p2
        return _sum_
    def minus(self):
        m = self.p1-self.p2
        return m
    def milt(self):
        mi = self.p1*self.p2
        return mi
    def dev(self):
        de = self.p1/self.p2
        return de
    def set_my_func(self, x_1,x_2 , method='method1' ):
        #method1 => y= self.p1*x_1+ self.p2* x_2     method2 => y=self.p1*x_1
        if method=='method1':
            y=self.p11 *x_1 + self.p22*x_2
            return y
        elif method=='method2':
            y=self.p1*x_1
            return y
```

```python
c = my_cla_funs(10,1) #在做init初始化
#調用class裡面的def
y = c.set_my_func(5,2,method='method2') 
# y=10*5+1*2
y
```

```python
#開始把畫圖包成class

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

class drew_picture:
    def __init__(self,file_path , target_column = '收盤價'): #賦予基礎技能,變數
        self.path = file_path  #我要加self 的目的是要得到等號右邊的變數
        self.column = target_column
        df = pd.read_csv(self.path, encoding= 'utf-8')
        self.columns = df.columns
        my_data = df[self.column].values.tolist()
        self.data = my_data
    #開始寫細部def
    def plot(self,title = 'my_plot'):
        x = range(len(self.data))
        y = self.data
        plt.title(title)
        plt.plot(x,y)
        plt.show()
        
    def scatter(self):
        x = range(len(self.data))
        y = self.data
        plt.scatter(x,y)
        plt.show()
```

```python
D = drew_picture('./stock.csv',target_column = '收盤價')
D.plot(title = '222222')
D.scatter()
```
