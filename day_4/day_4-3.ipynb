```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
my_font = fm.FontProperties(fname='./kaiu.ttf')

#畫圖, 中文亂碼問題
#win10用戶
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
# mac 用戶
# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
#開始畫圖  如果只畫一張圖, 要做細部註解

path  = './stock.csv'
df = pd.read_csv(path,encoding = 'utf-8')
#======針對收盤價作圖=================
close = df['收盤價'].values.tolist()
x_c = range(len(close))
y_c = close
y_c_max = np.array(y_c).max()
idx = y_c.index(y_c_max)
print('最大值的座標:', idx , y_c_max)

#======針對交易量作圖=================
def my_filter(data):  # data=628
    if type(data)==str:
        data = data.replace(',','')
        data = int(data)
        return data
    else:
        return data
df['成交筆數'] = df['成交筆數'].apply(my_filter)    
lot   = df['成交筆數'].values.tolist()
x_l = range(len(lot))
y_l = lot
#========================================
plt.plot(x_c,y_c,label = '收盤價')
plt.plot(x_l,y_l,label = '成交量')
plt.xlabel('天數')
plt.ylabel('交易量')
plt.title('交易內容')
plt.annotate(f'最大值:{y_c_max}',xy=(idx , y_c_max))
plt.legend()
plt.show()
plt.clf()
```

```python
# 畫多圖的話
fig = plt.figure(figsize=(5,7)) #準備畫布
#準備子圖
ax1 = plt.subplot(211)   #三位數: y有幾個圖 x有幾個圖  第幾張圖
ax2 = plt.subplot(212)
#===========開始話第一張圖======================
ax1.plot(x_c,y_c,label = '收盤價')
ax1.set_xlabel('天數')
ax1.set_ylabel('交易量')
ax1.set_title('交易內容')
ax1.annotate(f'最大值:{y_c_max}',xy=(idx , y_c_max))
ax1.legend()
#===========開始話第二張圖======================

ax2.plot(x_l,y_l,label = '成交量')
ax2.set_xlabel('天數')
ax2.set_ylabel('交易量')
ax2.set_title('交易內容')
ax2.legend()

plt.show()
```

```python
# 畫一個圖的,雙軸
fig = plt.figure(figsize=(5,7)) #準備畫布
#準備子圖
ax1 = plt.subplot(111)   #三位數: y有幾個圖 x有幾個圖  第幾張圖

#===========開始話第一張圖======================
ax1.plot(x_c,y_c,color = 'r',label = '收盤價')
ax1.set_xlabel('天數')
ax1.set_ylabel('收盤價')
ax1.set_title('交易內容')
ax1.annotate(f'最大值:{y_c_max}',xy=(idx , y_c_max))
ax1.legend()
#===========開始話第二張圖======================
ax1_2 = ax1.twinx()
ax1_2.plot(x_l,y_l,label = '成交量')
ax1_2.set_xlabel('天數')
ax1_2.set_ylabel('交易量')
ax1_2.set_title('交易內容')
ax1_2.legend()

plt.show()
```
