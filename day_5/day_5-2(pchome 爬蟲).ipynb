```python
import numpy as np
import pandas as pd
# !pip install requests
import requests #是一個工具包, 用來詢問前檯小姐姐問題的工具包
import json
```

```python
#step 1  把我們的問題列出來, 去得到小姐姐給的dict字典答案
url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone&page=1&sort=sale/dc' #我的問題
response = requests.get(url) #輸入問題給小姊姊,然後得到response
# info = response.text
# info
# res1 = json.loads(info) #json.loads 是把str 轉成python 看得懂的dict格式
# res1

#方法2 如果你的網址 不是utf8 做編碼, 就要使用下面代碼去針對不同編碼解碼
content = response.content.decode(encoding = 'utf-8')
res = json.loads(content)
res
```

```python
#step2 找一下我們要的內容發現是在prods的value中
prods=res['prods']
#發現prods 是一個list  [{第一個產品內容},{第二個產品內容}.....]
name_list        = []
price_list       = []
originPrice_list = []

for each_prod in prods:
    name        = each_prod['name']
    price       = each_prod['price'] 
    originPrice = each_prod['originPrice']
    name_list.append(name)
    price_list.append(price)
    originPrice_list.append(originPrice)
    
name_list    
```

```python
#step3 把籃子轉換成df 
df = pd.DataFrame(columns = ['name','price','originPrice']) #創一個空的df 去把list 灌回欄位中
df['name']= name_list
df['price']= price_list
df['originPrice']= originPrice_list
df
```

# 以上是爬取一個request / 準備開始變更url 的page 去爬取多頁內容

```python
#開始爬取多頁
import time
page=1
name_list        = []
price_list       = []
originPrice_list = []
df = pd.DataFrame(columns=['name','price','originPrice'])
while page<=10:
    url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone&page='+str(page)+'&sort=sale/dc' #我的問題
    response = requests.get(url) #輸入問題給小姊姊,然後得到response
    content = response.content.decode(encoding = 'utf-8')
    res = json.loads(content)
    prods=res['prods']
    print('目前爬取頁面:',page)
    print(url)
    for each_prod in prods:
        name        = each_prod['name']
        price       = each_prod['price'] 
        originPrice = each_prod['originPrice']
        name_list.append(name)
        price_list.append(price)
        originPrice_list.append(originPrice)
    page = page+1
    print('=====每頁爬取完畢=========','name籃子裡面的數量:',len(name_list))
    time.sleep(3)

df['name']= name_list
df['price']= price_list
df['originPrice']= originPrice_list
df
```

```python
#數據太亂太多, 我要針對某一個col 去做排序
df_sort =df.sort_values(by=['price'],na_position='first')
df_sort
```

```python
# 排序完之後發現有一些價格低的可能不是我想要的,所以用mask 去做過濾, 不符合的以NaN 空值取代
df_sort['price'] = df_sort['price'].mask(df_sort['price']<=15000 , np.NaN) 
#mask 其實跟apply 有點像, 只是apply(我自訂義的function), mask(我自己設定的條件)
df_sort
```

```python
# 雖然df 已經過濾完我的條件, 但是一堆NaN 不好看, 使用dropna直接刪掉
df_final = df_sort.dropna()
df_final.reset_index(drop=True ,  inplace=True) #重新排序依照index大小, 之前是nan的index 就會直接被遞補取代
df_final
#開始存檔 csv
df_final.to_csv('./df_final.csv',encoding = 'utf-8')
```

```python
# 假設 今天要針對name 去過濾 我只要是iphone 13 , 不要256g
def my_filter_keyword(s,keyword = 'iPhone 13' ,invkeyword = '256G' ):
    if keyword in s and invkeyword not in s:
        return s
    else:
        return np.NaN
df_final['name'] = df_final['name'].apply(my_filter_keyword) 
df_final
```
