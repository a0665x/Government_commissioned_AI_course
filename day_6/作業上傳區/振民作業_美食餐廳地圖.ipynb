```python
#爬一整頁
import requests#向前台要資料
from bs4 import BeautifulSoup

url='https://tabelog.com/tw/rstLst/japanese/'
my_headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'}
response = requests.get(url , headers = my_headers)
response.text
soup = BeautifulSoup(response.text , 'html.parser')
windose=soup.find_all('div',class_="lookup-result") #windose 是list 形式的beautifulsoup# bs4.element.Tag
windows_bs4 = windose[0]
boxs=windows_bs4.find_all('li',class_="list-rst js-list-item")

for box in boxs:
    box_img=box.find('a',class_="c-img-target")
    img_link=box_img.get('href')
    box_val=box.find('b',class_="c-rating__val")
    target_val=box_val.text
    box_price=box.find('span',class_="c-rating__val")
    target_price=box_price.text
    
    print('圖片連結:',img_link)
    print('評價:',target_val)
    print('售價:',target_price)
    print('=======爬取完成=========')
```

```python
#爬多頁
import requests#向前台要資料
from bs4 import BeautifulSoup
import time
import random

img_list  =[]
val_list  =[]
price_list=[]

page_from = 1
while True:
    try:
        rst= random.randint(5,15)
        url='https://tabelog.com/tw/tokyo/rstLst/'+str(page_from)+'/?LstCatD=RC0101&LstCat=RC01&Cat=RC'
        my_headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'}
        response = requests.get(url , headers = my_headers)
        response.text
        soup = BeautifulSoup(response.text , 'html.parser')
        windose=soup.find_all('div',class_="lookup-result") #windose 是list 形式的beautifulsoup# bs4.element.Tag
        windows_bs4 = windose[0]
        boxs=windows_bs4.find_all('li',class_="list-rst js-list-item")

        for idx,box in enumerate(boxs):
            box_img=box.find('a',class_="c-img-target")
            img_link=box_img.get('href')
            box_val=box.find('b',class_="c-rating__val")
            target_val=box_val.text
            box_price=box.find('span',class_="c-rating__val")
            target_price=box_price.text
            img_list.append(img_link)
            val_list.append(target_val)
            price_list.append( target_price)
            print(f'======這頁({page_from})內容爬取完成了,休息{rst}秒========')
        page_from = page_from+1
        time.sleep(rst)

    except:
        print('偵測到錯誤')
        #time.sleep(5)
        break
    
#     print('圖片連結:',img_link)
#     print('評價:',target_val)
#     print('售價:',target_price)
#     print('=======爬取完成=========')
```

```python
# #爬完資料,可以存到我們想要的資料夾
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

df = pd.DataFrame(columns = ['img_link','val','price'])
df['img_link'] = img_list
df['val']      = val_list
df['price']    = price_list
df

if not os.path.exists('./csv_file'):
    os.makedirs('./csv_file')
df.to_csv('./csv_file/Jananese food.csv')
```
