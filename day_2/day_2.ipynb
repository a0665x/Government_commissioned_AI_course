```python
#有限迴圈: for 每個元素 in 籃子:
#我要把籃子裡面的元素全部加100
data_list = [10,22,34,56,35,11,23,10,2]
cunt = 0
my_new_treat = []  #創造一個空得籃子, 目的是去接(append)每次循環的每個處理結果
for value in data_list:
#     print(f'正在處理第{cunt}次循環')
#     print('處理前的元素值',value)
    n = value+100
    my_new_treat.append(n)
    print('當下循環下,看一下添加的結果',my_new_treat)
    cunt = cunt+1  # 新的循環計數器 = 上一個循環的計數器 +1
    print(f'===處理完成===')
#用懶人寫法:
print('懶人寫法:')
print([i+100 for i in data_list])
print(my_new_treat)    
# my_new_treat
#學習重點:
#1. for in循環, 是針對一個籃子裡面的每個元素去做處理
#2. cunt 計數器, 可以應用在循環中, 每次循環,都會+1
#3. append 功能是list的其中一個小功能, 很常用在for in 循環中, 去把處理完的值,添加到這個新籃子內
#4. 用懶人寫法的話, 就不用額外去定義一個空的籃子去裝, 因為寫法裡面的最外圍就有一個[]
```
